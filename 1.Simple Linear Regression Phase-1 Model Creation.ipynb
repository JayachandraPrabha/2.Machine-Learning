Simple Linear Regression Phase-1 (Model Creation)
Steps:
1) Data Collection
2) Data preprocessing
3) Input & output split
4) Train & test split
5) Considering, Train set
6) Model creation
7) Considering, Test set
8) Evaluation Metrics
9) Cross check the best model
10) Save the best model

```python
# importing pandas library as pd & its all functions to perform all tabulation & database related tasks 
import pandas as pd
```

```python
# calling the read_csv function from pandas library inorder to retrive the database in the tabular form.
# Step 1) Data Collection
# Step 2) Data preprocessing

dataset = pd.read_csv("Salary_Data.csv")
dataset
```

```python
# Step 3) Input split
# Here the input (YearsExperience) represented as independent variable, inorder to split the input from the database named as dataset
independent = dataset[["YearsExperience"]]
independent
```

```python
# Step 3) output split
# Here the output (Salary) represented as dependent variable, inorder to split the output from the database named as dataset
dependent=dataset[["Salary"]]
dependent
```

```python
# Step 4) Train & test split
# from the file sklearn, opening the folder called model selection inorder to import train, test split function
from sklearn.model_selection import train_test_split
```

```python
# Inorder to use the train_test_split function, we basically need to pass 4 parameters like, input,output,test size & random state & 
# stored into 4 variables, so that which can be used in further functions
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.10,random_state=0)
x_train,x_test,y_train,y_test
```

```python
# Step 5) Considering, Train set
# 6) Model creation
# from the file sklearn, opening the folder called linear_model inorder to import Linear Regression function
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
# assigning all the parameters of the linear regression function into a variable called regressor
regressor.fit(x_train,y_train)
# The fit function is responsible for all the calculations of the Simple linear regression, 
# including the formula y=wx+b (w=Weight & b=bia), inorder to find the value of w & b 
# except x, where x will be substituted with the test value(x_test)
```

```python
# From the formula y=wx+b (w=Weight) is determined from the variable called 'coef_'
weight=regressor.coef_
weight
```

```python
# From the formula y=wx+b (b=bias) is determined from the variable called 'intercept_'
bias=regressor.intercept_
bias
```

```python
# step 7) Considering Test set
# For predicting the predicted value from the x_test i.e (independent test set) by assigning a variable called y_predict
y_predict=regressor.predict(x_test)
y_predict
```

```python
# step 8) Evaluation Metrics
# 9) Cross check the best model

# from the file sklearn, opening the folder called metrics inorder to import r-square(r2_score) function
from sklearn.metrics import r2_score
# Passing dependent test argument & predicted value (y_predict) in the r2_score function
r_score=r2_score(y_test,y_predict)
r_score
# Printed the value of r-square
# if r-square value is nearing 0 then it is poor model &
# if r-square value is nearing 1 then it is good model, Hence this is GOOD model
```

```python
# Step 10) Save the best model  
# importing the pickle library as pk inorder to save the obtained model as binary format
import pickle as pk
file_name="final_model_obtained.sav"
# Created a pickle file with .sav extension & stored in a variable called file_name
```

```python
# Storing the model into the pickle file with the help of the dump function by passing the arguments as regressor & 
# open function with file name in write binary mode
pk.dump(regressor,open(file_name,"wb"))
```

```python
# Obtaining the input from the user
user_data=int(input("enter the year: "))
# For loading the pickle file with the help of the load function by passing the open function with file nmae in read binary mode 
load_file=pk.load(open(file_name,"rb"))
result=load_file.predict([[user_data]])
# Storing the result in variable by providing (dot operator/access operator) to predict the user providing input
result
```
