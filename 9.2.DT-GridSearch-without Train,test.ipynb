```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male','smoker_yes']]
```

```python
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor 
grid_params={'criterion':['squared_error','friedman_mse','absolute_error','poisson'],
             'splitter':['best','random'],
             'max_features':['None','sqrt','log2','auto']
            }
grid=GridSearchCV(DecisionTreeRegressor(),grid_params,refit=True,verbose=3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
result=grid.cv_results_
# result
print("result is {}".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(result)
table
```

```python
'age', 'bmi', 'children', 'charges', 'sex_male', 'smoker_yes'
age=int(input("Enter the Age: "))
bmi=int(input("Enter the BMI_range: "))
children=int(input("Enter the Children_count: "))
sex_male=int(input("Enter the sex, male 1/female 0: "))
smoker_yes=int(input("Enter the smoker, yes 1/no 0): "))

pred=grid.predict([[age,bmi,children,sex_male,smoker_yes]])
print("Prediction is: {}".format(pred))
```
