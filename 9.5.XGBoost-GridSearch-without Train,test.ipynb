```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male','smoker_yes']]
```

```python
dependent=dataset[['charges']]
```

```python
import xgboost as xg
from xgboost import XGBRegressor
```

```python
from sklearn.model_selection import GridSearchCV
grid_params={"n_estimators":[100,500],
             "max_depth":[3,4,5],
             "learning_rate":[0.3,0.1]
            }
grid=GridSearchCV(XGBRegressor(),grid_params,refit=True,verbose=3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
# grid
```

```python
result=grid.cv_results_
print("The result is: {}".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(result)
table
```

```python
import pickle as pk
import xgboost as xg
```

```python
regressor=XGBRegressor(learning_rate= 0.1, max_depth= 3, n_estimators= 100)
regressor.fit(independent,dependent)

file_name="xg_boost_final_grid_model.sav"
pk.dump(regressor,open(file_name,"wb"))
```
